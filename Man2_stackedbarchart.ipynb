```python
import pandas as pd
import matplotlib.pyplot as plt

# Read Excel data
excel_file = '/content/Manuscript 2 Figures 2-3.xlsx'
sheet_name = 'Sheet1'
df = pd.read_excel(excel_file, sheet_name=sheet_name)

# Define the desired order of values
value_order = [
    'No Recall',
    'Unaided Recall',
    'Aided Recall'
]

# Reorder the columns
df = df[['Unnamed: 0'] + value_order]

# Set 'Unnamed: 0' as the index
df.set_index('Unnamed: 0', inplace=True)

# Convert values to percentages
df *= 100

# Create the horizontal stacked bar chart using matplotlib
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the data
bottom = None
for value in value_order:
    ax.barh(df.index, df[value], label=value, left=bottom)
    if bottom is None:
        bottom = df[value]
    else:
        bottom += df[value]

# Customize the chart
plt.title('Horizontal Stacked Bar Chart')
plt.xlabel('Percentage')
plt.ylabel('Categories')
plt.legend(title='Values')
ax.spines['top'].set_visible(False)  # Remove top spine
ax.spines['right'].set_visible(False)  # Remove right spine

# Format x-axis ticks as percentages
ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'{x:.0f}%'))

plt.tight_layout()  # Adjust layout

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set Seaborn style with grid turned off
sns.set(style="whitegrid", rc={"axes.grid": False})

# Read Excel data
excel_file = '/content/Manuscript 2 Figures 2-3.xlsx'
sheet_name = 'Sheet1'
df = pd.read_excel(excel_file, sheet_name=sheet_name)

# Define the desired order of values
value_order = [
    'Unaided Recall',
    'Aided Recall',
    'No Recall',
]

# Reorder the columns
df = df[['Unnamed: 0'] + value_order]

# Set 'Unnamed: 0' as the index
df.set_index('Unnamed: 0', inplace=True)

# Convert values to percentages
df *= 100

# Define custom colors
colors = ['mediumseagreen','steelblue', 'gray']

# Create the horizontal stacked bar chart using matplotlib
fig, ax = plt.subplots(figsize=(11, 7), dpi=400)  # Adjust DPI here

# Plot the data with custom colors
bottom = None
for value, color in zip(value_order, colors):
    bars = ax.barh(df.index, df[value], color=color, left=bottom, height=0.75, linewidth=0)
    for bar in bars:
        x_position = min(df[value]) - 3.5  # Position all labels on the right side
        width = bar.get_width()
        # Calculate dynamic x-position for percentage labels
        if width / max(df[value]) > 3:
            x_position = width + 3.2  # If bar is less than half, place label on the bar
        ax.text(x_position, bar.get_y() + bar.get_height()/2, f'{width:.1f}', va='center', fontsize=10, color='white', weight='heavy')
    if bottom is None:
        bottom = df[value]
    else:
        bottom += df[value]

# Customize the chart
plt.title('')
plt.xlabel('')
plt.ylabel('')
ax.spines['top'].set_visible(False)  # Remove top spine
ax.spines['right'].set_visible(False)  # Remove right spine
ax.spines['left'].set_visible(False)
ax.xaxis.set_visible(False)

# Format x-axis ticks as percentages
ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'{x:.0f}%'))

# Remove legend
ax.legend().set_visible(False)

plt.tight_layout()  # Adjust layout

# Save the plot with higher resolution
output_file = 'stacked_bar_chart.png'
plt.savefig(output_file, dpi=600)  # Save with higher DPI

# Show the plot
plt.show()
```
