```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import textwrap

from matplotlib import rcParams
rcParams['figure.dpi'] = 300

# Define your data matrix as a DataFrame
data = pd.DataFrame([
    [0.13, -0.11, 0.18, -0.02, -0.11, 0.16, 0.01, -0.01, 0.04, 0.04, 0.00, 0.01],
    [None, 0.06, 0.17, 0.18, 0.10, 0.39, 0.01, 0.02, -0.02, -0.06, 0.00, 0.04],
    [None, None, -0.08, 0.19, 0.40, -0.12, 0.04, 0.03, -0.08, -0.06, 0.02, 0.01],
    [None, None, None, 0.11, 0.02, 0.27, -0.15, 0.10, 0.05, -0.06, -0.03, 0.02],
    [None, None, None, None, 0.27, 0.10, 0.02, -0.02, 0.04, -0.06, 0.01, 0.00],
    [None, None, None, None, None, -0.08, 0.06, -0.05, -0.06, -0.14, -0.04, 0.06],
    [None, None, None, None, None, None, -0.11, 0.07, 0.04, 0.03, -0.05, 0.00]
])

# Add row and column labels
row_labels = ['1. Spending more time outdoors during a heat wave puts me at higher risk for heat stroke.', '2. Some heat waves are more dangerous than others.', "3. Healthy people don't need to change their daily routines during a heat wave.", '4. I know what to do to protect myself during a heat wave.', '5. Extreme heat is a health risk but manageable in the area I live.', '6. I am a person who does not need to worry in the event of a heat wave.', '7. The government should warn people about extreme heat events.']
column_labels = ['2. Some heat waves are more dangerous than others.', "3. Healthy people don't need to change their daily routines during a heat wave.", '4. I know what to do to protect myself during a heat wave.', '5. Extreme heat is a health risk but manageable in the area I live.', '6. I am a person who does not need to worry in the event of a heat wave.','7. The government should warn people about extreme heat events.', 'Age (24-44)', 'Age (60+)', 'Female', 'Poor/Fair Health', 'Low Income', 'High Income']

data.index = row_labels  # Assign row labels
data.columns = column_labels  # Assign column labels

display(data)

colors = ['red', 'white', 'blue']
cmap = mcolors.LinearSegmentedColormap.from_list('Custom', colors)

# Create a figure and axes
fig, ax = plt.subplots(figsize=(16, 9))

ax.set_facecolor('lightgray')

# Create the correlation matrix plot with custom annotations
heatmap = sns.heatmap(data, annot=True, fmt=".2f", cmap=cmap, cbar=True, vmin=-1, vmax=1, ax=ax)

# Adjust the y-axis tick labels and font size
plt.yticks(rotation=0, ha='right', fontsize=20)

# Rotate the x-axis tick labels for better readability
plt.xticks(rotation=45, ha='right', fontsize=20)

# Set the x-axis labels on top
ax.xaxis.set_ticks_position('top')

# Wrap the row labels on multiple lines
row_labels_multiline = [textwrap.fill(label, width=25) for label in row_labels]
heatmap.set_yticklabels(row_labels_multiline, rotation=0, ha='right', fontsize=12)

# Wrap the column labels on multiple lines
column_labels_multiline = [textwrap.fill(label, width=35) for label in column_labels]
heatmap.set_xticklabels(column_labels_multiline, rotation=45, ha='left', fontsize=12)

# Increase spacing between x-axis and y-axis labels
plt.tick_params(axis='x', which='minor', pad=40)  # Increase the pad value as desired

# Set the title and adjust the layout
plt.title('')
plt.tight_layout()

# Show the plot
plt.show()
```
