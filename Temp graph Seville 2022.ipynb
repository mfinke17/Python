```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = 'summer2022.csv'  # Use the actual uploaded filename

# Read the CSV file with the correct delimiter
data = pd.read_csv(file_path, delimiter=';')

# Convert the 'date' column to datetime format
data['date'] = pd.to_datetime(data['date'])

# Plotting code using the 'tmax' column
plt.plot(data['date'], data['tmax'])
plt.xlabel('Date')
plt.ylabel('Temperature (tmax)')
plt.title('Temperature Variation Over Time')
plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility
plt.tight_layout()  # Adjust layout to prevent label cutoff
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

file_path = 'summer2022.csv'  # Use the actual uploaded filename

# Read the CSV file with the correct delimiter
data = pd.read_csv(file_path, delimiter=';')

# Convert the 'date' column to datetime format
data['date'] = pd.to_datetime(data['date'])

# Set up figure and axes
plt.figure(figsize=(10, 6))
plt.style.use('seaborn')  # Apply a seaborn style

# Plotting code using the 'tmax' column
plt.plot(data['date'], data['tmax'], label='Max Temperature', color='red')

# Add labels and title
plt.xlabel('Date')
plt.ylabel('Temperature (C)')
plt.title(' ')

# Format x-axis ticks
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%b %d'))
plt.gca().xaxis.set_major_locator(mdates.DayLocator(interval=7))  # Show ticks every 7 days

# Rotate x-axis labels for better visibility
plt.xticks(rotation=45)

# Add a legend
plt.legend()

# Add padding to the layout
plt.tight_layout()

# Show the plot
plt.show()
```
